## 1) ¿Qué es un ciclo?
Un **ciclo** repite un bloque de instrucciones varias veces.

En Python tenemos principalmente:
- `for`: cuando sabemos qué elementos queremos recorrer.
- `while`: cuando repetimos hasta que una condición deje de cumplirse.

## 2) Ciclo `for`
Usamos `for` para recorrer listas, cadenas o rangos de números (`range`).

**Estructura básica:**

```python
for elemento in secuencia:
    # bloque a repetir
```

````python
# Ejemplo 1: recorrer una lista
frutas = ["manzana", "pera", "plátano"]
for fruta in frutas:
    print("Fruta:", fruta)

print("-" * 30)

# Ejemplo 2: usar range(inicio, fin)
for numero in range(1, 6):
    print("Número:", numero)

print("-" * 30)

# Ejemplo 3: acumular una suma
suma = 0
for numero in range(1, 6):
    suma += numero
print("Suma del 1 al 5:", suma)
````

## 3) Ciclo `while`
`while` repite un bloque **mientras** la condición sea verdadera.

**Estructura básica:**

```python
while condicion:
    # bloque a repetir
```

Recuerda actualizar variables dentro del ciclo para evitar ciclos infinitos.

````python
# Ejemplo 1: contar del 1 al 5
contador = 1
while contador <= 5:
    print("Contador:", contador)
    contador += 1

print("-" * 30)

# Ejemplo 2: break (detener ciclo)
numero = 1
while True:
    print("Número:", numero)
    if numero == 3:
        print("Se alcanzó el 3, saliendo con break")
        break
    numero += 1

print("-" * 30)

# Ejemplo 3: continue (saltar una iteración)
for n in range(1, 6):
    if n == 3:
        continue
    print("Valor:", n)
````
